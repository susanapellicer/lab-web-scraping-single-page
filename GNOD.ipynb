# Lab | Web Scraping Single Page (GNOD part 1)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 1. find url and store it in a variable
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
response = requests.get(url)
response.status_code
```

```python
#response.content
```

```python
len(response.content)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# soup
```

```python
soup.select("#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p")
```

```python
soup.select("p.title-artist")
```

```python
soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper p cite")
```

```python
soup.select("p.title-artist")[0]
soup.select("p.title-artist")[0].get_text()
```

```python
titles = soup.select("div.col-sm-8 p")

title_list = []

for title in titles:
    title_text = title.find(class_="title").get_text()
    title_list.append(title_text)

print(title_list)
```

```python
artist = soup.select("div.col-sm-8 p")

artist_list = []

for artist in artist:
    artist_text = artist.find(class_="artist").get_text()
    artist_list.append(artist_text)

print(artist_list)
```

```python
data = pd.DataFrame({"title":title_list,
                       "artist":artist_list
                      })
```

```python
data
```

```python
data.to_csv('top100songs.csv', index=False)
```

# Lab | Web Scraping Multiple Pages

### Instructions Part 1

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url2 = "https://en.wikipedia.org/wiki/List_of_rock_instrumentals"
```

```python
response2 = requests.get(url2)
response2.status_code
```

```python
soup2 = BeautifulSoup(response2.content, "html.parser")
```

```python
soup2
```

```python
soup2.select("#mw-content-text > div.mw-parser-output > table:nth-child(8)")
```

```python
# soup2.select("table")[0]
```

```python
prestab = soup2.select("table")[0]
prestab
prestab.select("tr td a")
```

```python
soup2.select("#mw-content-text > div.mw-parser-output > table:nth-child(8) a")[0].title
#soup2.select("#mw-content-text > div.mw-parser-output > table:nth-child(8) a") #[0].get_text()
```

```python
artists = []
song_titles = []

for row in prestab.select("tr"):
    cells = row.select("td")
    if len(cells) >= 2:
        artist = cells[0].text.strip()
        song_title = cells[1].text.strip()
        artists.append(artist)
        song_titles.append(song_title)
```

```python
artists
```

```python
song_titles
```

```python
data1 = {"Artist": artists, "Song Title": song_titles}
data1 = pd.DataFrame(data1)
data1
```

```python
data1.to_csv('top100songs.csv', index=False)
```
