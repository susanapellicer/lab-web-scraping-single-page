# Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page: url ='https://en.wikipedia.org/wiki/Python'

```python
url ='https://en.wikipedia.org/wiki/Python'
```

```python
links = requests.get(url)
links.status_code
```

```python
soup = BeautifulSoup(links.content, "html.parser")
soup
```

```python
#Option 1:

links = []
for link in soup.findAll('a'):
    links.append(link.get('href'))

links
```

```python
len(links)
```

```python
#Option 2:

from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re

req = Request("https://en.wikipedia.org/wiki/Python")
html_page = urlopen(req)

soup2 = BeautifulSoup(html_page, "lxml")

links2 = []
for link in soup2.findAll('a'):
    links2.append(link.get('href'))

links2
```

# Display the top 10 languages by number of native speakers stored in a pandas dataframe: url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

```python
url ='https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
top_languages = requests.get(url)
top_languages.status_code
```

```python
soup5 = BeautifulSoup(top_languages.content, "html.parser")
#soup5
```

```python
#soup5.select('#mw-content-text > div.mw-parser-output ')
```

```python
languages = soup5.select("td > a")

top_languages_list = []

for title in languages:
    title_text = title.get_text()
    top_languages_list.append(title_text)
    
top_languages_list[-10:]
```
